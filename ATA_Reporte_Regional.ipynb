```python
import pandas as pd
import numpy as np
import mysql.connector as connection

host="localhost"
database ='difoca'
user="root"
passwd="root1"
```

```python
try:
    con = connection.connect(host="localhost", database ='difoca',user="root", passwd="root1",use_pure=True)
    query = "select * from  v_ata_resumen"
    query2 = "select * from v_ata_participacion_po_encuesta"
    query3 = "select * from ma_iged"
    df1 = pd.read_sql(query,con)
    df2 = pd.read_sql(query2,con)
    df3 = pd.read_sql(query3,con)
    con.close()
except Exception as e:
    con.close()
    print(str(e))

df2["llave1"]=df2["cod_ata"].map(str)+df2["sesion"].map(str)


#Definir las regiones y/o macro
#region
reg="TUMBES"
df2 = df2[(df2["regionIGED"]==reg)]
mapa = "C:/Users/difoca05/Pictures/Regiones/"+reg+".png"
kpi_df3 = df3[(df3["REGION"]==reg)]
kpi_5 = kpi_df3["NOMBRE_IGED"].count()

#macro
# df2 = df2[(df2["regionIGED"]=="ANCASH")|(df2["regionIGED"]=="CAJAMARCA")|(df2["regionIGED"]=="LA LIBERTAD")|(df2["regionIGED"]=="LAMBAYEQUE")|(df2["regionIGED"]=="PIURA")|(df2["regionIGED"]=="TUMBES")]
# kpi_df3 = df3[(df3["REGION"]=="ANCASH")|(df3["REGION"]=="CAJAMARCA")|(df3["REGION"]=="LA LIBERTAD")|(df3["REGION"]=="LAMBAYEQUE")|(df3["REGION"]=="PIURA")|(df3["REGION"]=="TUMBES")]
# kpi_5 = kpi_df3["NOMBRE_IGED"].count()
# mapa = "C:/Users/difoca05/Pictures/Regiones/"+"CAJAMARCA"+".png"


# 'TACNA', 'AREQUIPA','PUNO','CUSCO', 'MOQUEGUA'


df1_1 = pd.DataFrame(df2.groupby(["enfasis"])["cod_ata"].nunique())
df1_1.reset_index(inplace=True)
df1_1.sort_values(by=['cod_ata'], inplace=True, ascending=False)
```

```python
dfx = pd.DataFrame(df2.groupby('regionIGED').cod_ata.nunique())
dfx.reset_index(inplace=True)
dfx.sort_values(by=['cod_ata'], inplace=True, ascending=False)

dfy = pd.DataFrame(df1.groupby('mes').kpi.sum())
dfy.reset_index(inplace=True)

df1_2 = pd.DataFrame(df2.groupby(["uo"])["cod_ata"].nunique())
df1_2.reset_index(inplace=True)
df1_2.sort_values(by=['cod_ata'], inplace=True, ascending=False)

df2.fillna('', inplace=True)
df2[df2["ambito"]!=""]
df2["COD_IGED"]=df2["cod_iged"]
```

```python
dfx1 = pd.DataFrame(df2[df2["tipo_inst"]=="DRE/GRE"].groupby('nombre_iged').cod_ata.nunique())
dfx1.reset_index(inplace=True)
dfx1.sort_values(by=['cod_ata'], inplace=True, ascending=True)
dfx1.info()

dfx2 = pd.DataFrame(df2[df2["tipo_inst"]=="UGEL"].groupby('nombre_iged').cod_ata.nunique())
dfx2.reset_index(inplace=True)
dfx2.sort_values(by=['cod_ata'], inplace=True, ascending=True)
dfx2.info()

dfx3 = pd.DataFrame(df2[df2["ambito"]!=""].groupby(['nombre_iged', 'ambito', "COD_IGED"]).cod_ata.nunique())
dfx3.reset_index(inplace=True)
dfx3.sort_values(by=['cod_ata'], inplace=True, ascending=True)
df33=df3[["COD_IGED", "AMBITO"]]
dfx3=dfx3.merge(df33, on="COD_IGED", how="left")
```

```python
df22=df2[["nombre_iged", "COD_IGED", "cod_ata"]]
df22.drop(df22[df22['COD_IGED'] ==""].index, inplace = True)

df33=df3[["COD_IGED", "AMBITO", "Tipo_UGEL_2021"]]

dfx4 = pd.DataFrame(df22.groupby(['nombre_iged', "COD_IGED"]).cod_ata.nunique())
dfx4.reset_index(inplace=True)
dfx4.sort_values(by=['cod_ata'], inplace=True, ascending=True)
dfx4=dfx4.merge(df33, on="COD_IGED", how="left")
dfx4.fillna("", inplace=True)
dfx4=dfx4[dfx4["Tipo_UGEL_2021"]!=""]
```

```python
dfx5 = pd.DataFrame(df2.groupby(['titulo', 'enfasis']).po.sum())
dfx5.reset_index(inplace=True)
dfx5=dfx5[(dfx5["enfasis"]=="INSTITUCIONAL")|(dfx5["enfasis"]=="INSTITUCIONAL, ADMINISTRATIVA")].sort_values(by=['po'])

dfx6 = pd.DataFrame(df2.groupby(['titulo', 'enfasis']).po.sum())
dfx6.reset_index(inplace=True)
dfx6=dfx6[dfx6["enfasis"]=="PEDAGÓGICA"].sort_values(by=['po'])

dfx7 = pd.DataFrame(df2.groupby(['titulo', 'enfasis']).po.sum())
dfx7.reset_index(inplace=True)
dfx7=dfx7[(dfx7["enfasis"]=="ADMINISTRATIVO")|(dfx7["enfasis"]=="INSTITUCIONAL, ADMINISTRATIVA")].sort_values(by=['po'])
```

```python
df_a2=df2.drop(df2[df2['COD_IGED'] ==""].index)
dfx8 = pd.DataFrame(df_a2.groupby(['titulo', 'enfasis']).cod_iged.nunique())
dfx8.reset_index(inplace=True)
dfx8=dfx8[(dfx8["enfasis"]=="INSTITUCIONAL")|(dfx8["enfasis"]=="INSTITUCIONAL, ADMINISTRATIVA")|(dfx8["enfasis"]=="INSTITUCIONAL, PEDAGÓGICA")].sort_values(by=['cod_iged'])

dfx9 = pd.DataFrame(df_a2.groupby(['titulo', 'enfasis']).cod_iged.nunique())
dfx9.reset_index(inplace=True)
dfx9=dfx9[(dfx9["enfasis"]=="PEDAGÓGICA")|(dfx9["enfasis"]=="INSTITUCIONAL, PEDAGÓGICA")].sort_values(by=['cod_iged'])

dfx10 = pd.DataFrame(df_a2.groupby(['titulo', 'enfasis']).cod_iged.nunique())
dfx10.reset_index(inplace=True)
dfx10=dfx10[(dfx10["enfasis"]=="ADMINISTRATIVO")|(dfx10["enfasis"]=="INSTITUCIONAL, ADMINISTRATIVA")].sort_values(by=['cod_iged'])
dfx10
```
