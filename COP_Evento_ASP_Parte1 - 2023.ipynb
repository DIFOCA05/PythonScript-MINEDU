```python
#Conex Excel Sharepoint

from office365.runtime.auth.authentication_context import AuthenticationContext
from office365.sharepoint.client_context import ClientContext
from office365.sharepoint.files.file import File 
import io
import pandas as pd
from datetime import datetime
import mysql.connector as connection

host="localhost"
database ='difoca'
user="root"
passwd="root1"

username = "difoca05@minedu.gob.pe"
password = "Minedu2023++@"

#**********
#target url taken from sharepoint and credentials
evento = "ASESORIA PERSONALIZADA"
año ='2023'
url = 'https://mineduperu-my.sharepoint.com/personal/difoca05_minedu_gob_pe/Documents/COMUNIDAD%20DE%20PRACTICA%20(COP)/2023/EVENTOS/ASESORIA%20PERSONALIZADA/ASESORIA_PERSONALIZADA_2023.xlsx?web=1'
hoja = "ASESORIA_PERSONALIZADA"

ctx_auth = AuthenticationContext(url)
if ctx_auth.acquire_token_for_user(username, password):  
  ctx = ClientContext(url, ctx_auth)
  web = ctx.web
  ctx.load(web)
  ctx.execute_query()
  print("Authentication successful")

response = File.open_binary(ctx, url)

#save data to BytesIO stream
bytes_file_obj = io.BytesIO()
bytes_file_obj.write(response.content)
bytes_file_obj.seek(0) #set file object to start

df = pd.read_excel(bytes_file_obj, sheet_name=hoja)
#**********

df = df[["cod_evento", "periodo","Región","IGED","Nombre de la iniciativa","cod_ini","Registro en el Muro COP","Tema solicitado","Persona de contacto","Experto/Especialista","Fecha","Hora","Estado","N° participantes conectados","N° participantes registrados","Observaciones","kpi","Link de registros de asistencia-Respuestas"]]

# Limpieza sin código
df["cod_evento"].fillna('', inplace=True)
df = df.drop(df[df["cod_evento"]==""].index)
# Limpieza sin fecha
df["Fecha"].fillna('', inplace=True)
df = df.drop(df[df["Fecha"]==""].index)
df["Fecha"] = pd.to_datetime(df["Fecha"],format='%d-%m-%Y')

df1=df

df1["Fecha"] = df1["Fecha"].astype(str)
df1["N° participantes conectados"] = df1["N° participantes conectados"].astype(str)
df1["kpi"].fillna('0', inplace=True)
df1["kpi"] = df1["kpi"].astype(str)
df1
```
