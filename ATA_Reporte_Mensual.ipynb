```python
import pandas as pd
import numpy as np
import mysql.connector as connection
import openpyxl

host="localhost"
database ='difoca'
user="root"
passwd="root1"

mes = "11. Nov"
año = "2022"

try:
    con = connection.connect(host=host, database =database,user=user, passwd=passwd,use_pure=True)
    query = "select * from v_ata_cierre"
    df = pd.read_sql(query,con)
    con.close() #close the connection
except Exception as e:
    con.close()
    print(str(e))

excel = "C:/Users/difoca05/OneDrive - Ministerio de Educación/ASISTENCIA TECNICA (ATA)/Reporte ATA "+ mes[-3:] +" "+año+".xlsx"
df.to_excel(excel, index=False, sheet_name='base_datos')

df['uo'] = df['uo'].str.replace(' ','')
df1 = df.dropna(subset=["cod_ata1"])
df1.info()
```

```python
x1 = df["cod_sisged"][df["año"]==año].count() #ata regist año
x2 = df["cod_sisged"][(df['año'] == año) & (df['mes'] == mes)].count() #ata regist mes
x3 = df["uo"][df["año"]==año].nunique() #uo regist año
x4 = df["uo"][(df['año'] == año) & (df['mes'] == mes)].nunique() #uo regist mes

x5 = df1["cod_sisged"][df1["año"]==año].count() #ata imp año
x6 = df1["cod_sisged"][(df1["año"]==año) & (df1['mes'] == mes)].count() #ata imp mes
x7 = df1["uo"][df1["año"]==año].nunique() #uo imp año
x8 = df1["uo"][(df1['año'] == año) & (df1['mes'] == mes)].nunique() #uo imp mes
x9 = df1["sesion1"][df1["año"]==año].sum() #sesiones imp año
x10 = df1["sesion1"][(df1["año"]==año) & (df1['mes'] == mes)].sum() #sesiones imp mes
```

```python
y1 = df[df["año"]==año].groupby("uo")["cod_sisged"].nunique()
y2 = df1[df1["año"]==año].groupby("uo")["cod_sisged"].nunique()

dfy1 = pd.DataFrame(y1)
dfy1["U.O."] = dfy1.index
dfy1=dfy1.rename(columns={'cod_sisged':'ATA Registrada'})

dfy2 = pd.DataFrame(y2)
dfy2["U.O."] = dfy2.index
dfy2=dfy2.rename(columns={'cod_sisged':'ATA Implementada'})

dfy = dfy1.merge(dfy2, on="U.O.", how="left")
dfy=dfy[["U.O.", "ATA Registrada", "ATA Implementada"]]
dfy.fillna(0, inplace=True)
dfy["ATA Implementada"]=dfy["ATA Implementada"].astype(int)
dfy1
```

```python
z1 = df[(df['año'] == año) & (df['mes'] == mes)].groupby("uo")["cod_sisged"].nunique()
z2 = df1[(df1['año'] == año) & (df1['mes'] == mes)].groupby("uo")["cod_sisged"].nunique()

dfz1 = pd.DataFrame(z1)
dfz1["U.O."] = dfz1.index
dfz1=dfz1.rename(columns={'cod_sisged':'ATA Registrada'})

dfz2 = pd.DataFrame(z2)
dfz2["U.O."] = dfz2.index
dfz2=dfz2.rename(columns={'cod_sisged':'ATA Implementada'})

dfz = dfz1.merge(dfz2, on="U.O.", how="left")
dfz=dfz[["U.O.", "ATA Registrada", "ATA Implementada"]]
dfz.fillna(0, inplace=True)
dfz["ATA Implementada"]=dfz["ATA Implementada"].astype(int)
dfz #uo por mes
```
